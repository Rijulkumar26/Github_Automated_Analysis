```python
#Github url
github_url = input("Please enter Github URL: ")
```

```python
import requests

def fetch_user_repositories(github_url):
    # Extract username from the GitHub URL
    username = github_url.split('/')[-1]

    # Construct the API endpoint URL
    api_url = f"https://api.github.com/users/{username}/repos"
    

    # Send GET request to the GitHub API
    response = requests.get(api_url)

    if response.status_code == 200:
        # Parse the JSON response
        repositories = response.json()

        # Process the repository information as needed
        for repo in repositories:
            repo_name = repo["name"]
            repo_url = repo["html_url"]
            repository_urls.append(repo_url)
            print(f"Repository: {repo_name}\nURL: {repo_url}")
    else:
        print(f"Failed to fetch repositories: {response.status_code} - {response.text}")

repository_urls = []

# Call the function to fetch user repositories
fetch_user_repositories(github_url)
```

```python
import os
import requests
import nltk
from nltk.tokenize import word_tokenize
import pandas as pd


file_extensions = [".py", ".ipynb", ".md"]

data = []  # List to store the tokenized data


for url in repository_urls:
    api_url = url.replace("github.com", "api.github.com/repos")
    response = requests.get(api_url)

    if response.status_code == 200:
        repo_data = response.json()

        # Extract the username and repository name from the URL
        parts = url.split("/")
        username = parts[-2]
        repository_name = parts[-1]

        # Create a folder for the repository if it doesn't exist
        folder_path = f"{username}/{repository_name}"
        os.makedirs(folder_path, exist_ok=True)

        for extension in file_extensions:
            # Get the contents of the repository directory
            dir_url = f"{api_url}/contents"
            dir_response = requests.get(dir_url)

            if dir_response.status_code == 200:
                dir_data = dir_response.json()

                for item in dir_data:
                    if item["type"] == "file" and item["name"].endswith(extension):
                        file_url = item["download_url"]

                        file_response = requests.get(file_url)
                        if file_response.status_code == 200:
                            file_content = file_response.text

                            # Extract the filename from the file URL
                            file_name = item["name"]

                            # Path of all the files
                            file_path = f"{folder_path}/{file_name}"

                            # Tokenize the file content
                            tokens = word_tokenize(file_content)

                            # Add the tokenized data to the list
                            data.append((repository_name, tokens))

                            print(f"Successfully tokenized {file_path}")
                        else:
                            print(f"Failed to extract {file_url}")
            else:
                print(f"Failed to access repository contents: {dir_url}")
    else:
        print(f"Failed to access repository: {url}")

# Create a DataFrame with the tokenized data
df = pd.DataFrame(data, columns=["Repository", "Tokens"])
```

```python
import nltk
from nltk.corpus import stopwords
import string

# Remove stopwords and punctuation from the tokens in the DataFrame
stopwords_set = set(stopwords.words('english'))
punctuation_set = set(string.punctuation)

# Function to remove stopwords and punctuation from a list of tokens
def remove_stopwords_punctuation(tokens):
    return [token for token in tokens if token.lower() not in stopwords_set and token not in punctuation_set]

# Apply the function to remove stopwords and punctuation to the 'Tokens' column in the DataFrame
df['Tokens'] = df['Tokens'].apply(remove_stopwords_punctuation)
```

```python
import numpy as np

# Calculate metrics for each repository
df['Total Tokens'] = df['Tokens'].apply(len)
df['unique_token_count'] = df['Tokens'].apply(lambda x: len(set(x)))
df['Average Token Length'] = df['Tokens'].apply(lambda x: np.mean([len(token) for token in x]))

# Calculate complexity score based on the metrics
df['Complexity Score'] = df['Total Tokens'] * df['unique_token_count'] * df['Average Token Length']

df
```

```python
# Find the repository with the highest complexity score
most_complex_repository = df.loc[df['Complexity Score'].idxmax(), 'Repository']

print(f"The most complex repository is " + "\033[1m" + most_complex_repository + "\033[0m" + " based on the 3 factors of files in each repository - Size of the files and uniqueness of file")
```
